```python
import pandas as pd
import numpy as np

# 1. GENERACIÓN DE DATOS SIMULADOS (Como si vinieran de un PLC)
# Creamos un DataFrame con datos de 1 turno de 8 horas (480 min)
datos = {
    'hora': pd.date_range(start='2024-01-01 06:00', periods=480, freq='T'), # Cada minuto
    'estado_maquina': np.random.choice(['Operando', 'Falla', 'Espera_Material'], 480, p=[0.85, 0.10, 0.05]),
    'piezas_producidas': np.random.choice([0, 10, 12], 480, p=[0.15, 0.60, 0.25]), # 0 si está parada, 10-12 si opera
    'piezas_defectuosas': np.random.choice([0, 1, 2], 480, p=[0.95, 0.04, 0.01])
}
df = pd.DataFrame(datos)

# 2. CÁLCULO DE KPIS
# A. Disponibilidad: Tiempo Real Operando / Tiempo Planificado
tiempo_planificado = 480 # minutos
tiempo_operativo = df[df['estado_maquina'] == 'Operando'].shape[0]
disponibilidad = tiempo_operativo / tiempo_planificado

# B. Calidad: (Total - Defectos) / Total
total_piezas = df['piezas_producidas'].sum()
total_defectos = df['piezas_defectuosas'].sum()
piezas_buenas = total_piezas - total_defectos
calidad = piezas_buenas / total_piezas

# C. Rendimiento: (Total Piezas / Tiempo Operativo) / Capacidad Máxima Teórica
capacidad_max_teorica_por_min = 12 # La máquina da max 12 piezas/min
rendimiento = (total_piezas / tiempo_operativo) / capacidad_max_teorica_por_min

# 3. RESULTADO OEE
oee = disponibilidad * rendimiento * calidad

print(f"--- REPORTE DE PRODUCCIÓN ---")
print(f"Disponibilidad: {disponibilidad:.2%}")
print(f"Rendimiento:    {rendimiento:.2%}")
print(f"Calidad:        {calidad:.2%}")
print(f"-----------------------------")
print(f"OEE FINAL:      {oee:.2%}")

# Tip: Si el OEE < 65%, tu proceso necesita mejora urgente.
```
